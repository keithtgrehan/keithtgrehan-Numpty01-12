# Solutions: Intro to NumPy

```python
import numpy as np
```

Timing sum: NumPy vs list

```python
numpy_array = np.arange(0, 1000000)
python_list = list(range(1000000))
%timeit np.sum(numpy_array)
%timeit sum(python_list)
```

Create NumPy arrays

```python
my_lst_ndarray = np.array([1, 2, 3, 4, 5])
my_tuple_ndarray = np.array((1, 2, 3, 4, 5), np.int32)
```

```python
print(my_lst_ndarray.shape)
print(my_lst_ndarray.dtype)
```

```python
my_lst_ndarray2 = np.array(["1", 2, 3, "10", 5])
print(my_lst_ndarray2.dtype)  # Expect <U2 (Unicode string)
```

```python
my_lst_ndarray3 = np.array([1, 2, 3, "10", 5], np.int32)
print(my_lst_ndarray3.dtype)  # Casts to int32
```

Other array creation (zeros, ones, arange, linspace, eye)

```python
np.zeros((2, 3))
np.ones((2, 3))
np.arange(0, 10, 2)
np.linspace(0, 10, 6)
np.eye(3)
```

Reshape, indexing, slicing, views/copies

```python
range_arr = np.arange(0, 20).reshape((5, 4))
range_arr
```

```python
range_arr[2:4, 1:3]  # Slicing
```

Math operations

```python
range_arr + 2
range_arr * 3
range_arr ** 2
```

Stats (mean, std, max, min, etc.)

```python
range_arr.mean(axis=0)
range_arr.std(axis=0)
range_arr.max(axis=0)
range_arr.min(axis=0)
range_arr.argmin(axis=0)
range_arr.argmax(axis=0)
```

```python
range_arr.cumsum(axis=0)
range_arr.cumprod(axis=0)
```

```python
range_arr.flatten()
```

np.where example

```python
my_ndarray = np.array([2, 4, 6, 8, 24, 3, 8, 9, 12])
np.where(my_ndarray <= 2)
np.where(my_ndarray == 8)
np.where(my_ndarray > 6)
```

Pivot table example (from end of notebook)

```python
import pandas as pd
red_wines_df = pd.read_csv('data/winequality-red.csv', sep=';')
fixed_acidity_bins = np.arange(4, 17)
fixed_acidity_series = pd.cut(red_wines_df['fixed acidity'], bins=fixed_acidity_bins, labels=fixed_acidity_bins[:-1])
fixed_acidity_series.name = 'fa_bin'
red_wines_df = pd.concat([red_wines_df, fixed_acidity_series], axis=1)
pd.pivot_table(red_wines_df, values='residual sugar', index='quality', columns='fa_bin')
```
