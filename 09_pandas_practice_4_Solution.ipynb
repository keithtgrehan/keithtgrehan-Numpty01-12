3) I've decided that this month I want to stay away from wines with relatively high alcohol content. To do that, I'm going to avoid any wines that have a greater alcohol content than the mean alcohol content, and you're going to help me do this. To achieve this, let's do the following: 

  * Find the mean alcohol content, separately, for reds and whites.  
  * Create a `Series` that holds whether each row in each `DataFrame` (the reds, and whites) has a higher alcohol content than the mean. 
  * Merge this `Series` onto the `DataFrame`. I can imagine doing this with either a `.join()` or using `pd.concat()`. For practice, do it with both. Note: merges with `Series` work the same way that they work with `DataFrames`.  
  * Return back to me all those rows that will help me stay away from those wines with a higher alcohol content.  
   

```python
# 3a. Find the mean alcohol content, separately, for reds and whites.
print(f"Mean alcohol red wines: {wine_red.alcohol.mean()}")
print(f"Mean alcohol white wines: {wine_white.alcohol.mean()}")
```

```python
# 3b. Create a Series that holds whether each row in each DataFrame (the reds, and whites) 
# has a higher alcohol content than the mean.
alcohol_series_red = wine_red.alcohol > wine_red.alcohol.mean()
alcohol_series_white = wine_white.alcohol > wine_white.alcohol.mean()

# Rename series
alcohol_series_red.rename('alcohol_above_average', inplace=True)
alcohol_series_white.rename('alcohol_above_average', inplace=True)
```

```python
# 3c. Merge this Series onto the DataFrame.

# Using pd.concat()
#wine_red = pd.concat([wine_red, alcohol_series_red], axis=1)
#wine_white = pd.concat([wine_white, alcohol_series_white], axis=1)

# Using .join()
wine_red = wine_red.join(alcohol_series_red)
wine_white = wine_white.join(alcohol_series_white)

wine_red.head(2)
```

```python
# 3d. Return back to me all those rows that will help me stay away from those wines with a higher alcohol content.
wine_red_low_alcohol = wine_red[wine_red['alcohol_above_average'] == False]
wine_red_low_alcohol.head(2)
```

```python
wine_white_low_alcohol = wine_white.query('alcohol_above_average == False')
wine_white_low_alcohol.head(2)
```

4) Let's say that I want to get started on cutting back next month. This time, though, I want to focus on staying away from those wines with a high acidity. Specifically, I want to stay away from those wines that are in the highest bin of fixed acidity (highest bin out of 5). You're now going to help me do this. To achieve this, let's do the following: 

 * Separate the rows in each `DataFrame` into 5 equal width bins (not equal to quintiles) based off their fixed acidity. 
 * Merge the resulting `Series` holding these 5 bins onto the original `DataFrame`. I can imagine also doing this with either `.join()` or using `pd.concat()`. Try doing it with both for practice. 
 * Return back to me all those rows that are **not in** the top bin in terms of fixed acidity. 

```python
# 4a. Separate the rows in each DataFrame into 5 equal width bins (not equal to quintiles) based off their fixed acidity.
red_bins = pd.cut(wine_red['fixed acidity'], 5, labels=[1,2,3,4,5])
white_bins = pd.cut(wine_white['fixed acidity'], 5, labels=[1,2,3,4,5])

# rename Pandas series
red_bins.rename('acidity_bins', inplace=True)
white_bins.rename('acidity_bins', inplace=True)
```

```python
# 4b. Merge the resulting Series holding these 5 bins onto the original DataFrame.

# Using pd.concat()
#wine_red = pd.concat([wine_red, red_bins], axis=1)
#wine_white = pd.concat([wine_white, white_bins], axis=1)

# Using .join()
wine_red = wine_red.join(red_bins)
wine_white = wine_white.join(white_bins)

wine_red.head(2)
```

```python
# 4c. Return back to me all those rows that are not in the top bin in terms of fixed acidity.
wine_red_low_acidity = wine_red[wine_red['acidity_bins'] != 5]
wine_white_low_acidity = wine_white[wine_white['acidity_bins'] != 5]

wine_red_low_acidity.head(2)
```

5) Let's say that I now want to know how much my decision to avoid those wines with higher `alcohol` content is going to limit the `quality` of wines that I can drink. To figure this out, I want to know a couple of things: 

 * The average `alcohol` content for those reds above the mean `alcohol` level, by quality.
 * The average `alcohol` content for those whites above the mean `alcohol` level, by quality. 
 
 Use a `pivot table` to solve this. 

```python
# Red wines
pd.pivot_table(wine_red, values='alcohol', index='alcohol_above_average', columns='quality')
```

```python
# White wines (this time we can also use .pivot_tables())
wine_white.pivot_table(values='alcohol', index='alcohol_above_average', columns='quality')
```
