## Part 2: Manipulate NumPy arrays.

1. Please create a NumPy array with (1, 3, '4', 7, 12, '0'). Define the content as integer.


`````python
import numpy as np

# 1. Create a NumPy array with (1, 3, '4', 7, 12, '0') as integers
arr = np.array([1, 3, '4', 7, 12, '0'], dtype=int)
print("1) Initial array:", arr)
`````

2. Check the data type of the objects and the shape of the array


`````python
print("2) Data type:", arr.dtype)
print("   Shape:", arr.shape)
`````

3. Update the 4th value to 30.


`````python
arr[3] = 30
print("3) After updating 4th value to 30:", arr)
`````

4. Reshape the array to a 2x3 matrix.


`````python
arr2 = arr.reshape(2, 3)
print("4) Reshaped to 2x3:\n", arr2)
`````

5. Please add 8 to each value in the first row and 12 to each in the second row.


`````python
arr5 = arr2.copy()
arr5[0] += 8
arr5[1] += 12
print("5) After adding 8 to row 1, 12 to row 2:\n", arr5)
`````

6. Multiply the first column with 2, the second with 3 and the third with 4.


`````python
multipliers = np.array([2, 3, 4])
arr6 = arr5 * multipliers
print("6) After column-wise multiplication (2,3,4):\n", arr6)
`````

7. Please sum up all numbers in the first row, and all numbers in the second row.


`````python
row_sums = arr6.sum(axis=1)
print("7) Row sums:", row_sums)
`````

8. Similarly, search for the largest number for each column.


`````python
col_max = arr6.max(axis=0)
print("8) Max per column:", col_max)
`````

9. Extract the number in the second column and the first row.


`````python
value_2ndcol_1strow = arr6[0, 1]
print("9) Value at (row 0, col 1):", value_2ndcol_1strow)
`````

10. Check at which index the value is exactly 48.

`````python
indices_48 = np.where(arr6 == 48)
print("10) Indices where value == 48:", indices_48)
# If you want them as (row, col) pairs:
print("    As (row, col) list:", list(zip(indices_48[0], indices_48[1])))
`````
