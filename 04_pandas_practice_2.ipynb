```python
# import pandas
import pandas as pd
```

```python
# load the data
df = pd.read_csv('data/iris.csv')
```

Let us first have a look at the head of the table, maybe also on the last 10 rows...

```python
# 1) Look at the head and last 10 rows
print("HEAD:")
print(df.head())
print("\nTAIL (last 10 rows):")
print(df.tail(10))
```

How many irises are in the data set?

```python
# 2) How many irises are in the data set?
n_irises = len(df)
print("\nNumber of rows / irises:", n_irises)
```

How many different species are there?

```python
n_species = df["species"].nunique()
species_counts = df["species"].value_counts()
print("\nNumber of species:", n_species)
print("Counts per species:")
print(species_counts)
```

Calculate the mean, median, mode for petal length

```python
petal_len = df["petal_length"]

mean_petal_len = petal_len.mean()
median_petal_len = petal_len.median()
mode_petal_len = petal_len.mode()  # can be more than one value

print("\nPetal length stats:")
print("Mean:   ", mean_petal_len)
print("Median: ", median_petal_len)
print("Mode(s):")
print(mode_petal_len)
```

What can you conclude about the distribution for petal length?

What is the smallest and largest value for petal length?

```python
# 5) Smallest and largest petal length
min_petal_len = petal_len.min()
max_petal_len = petal_len.max()
print("\nMin petal length:", min_petal_len)
print("Max petal length:", max_petal_len)
```

Calculate the variance and standard deviation for petal length

```python
var_petal_len = petal_len.var()
std_petal_len = petal_len.std()
print("\nPetal length variance:", var_petal_len)
print("Petal length std dev:", std_petal_len)
```

Calculate the basic descriptive statistics for all columns of the iris data set with a single command.

```python
print("\nFull describe():")
print(df.describe())
```

What is the overall average for sepal length?

```python
avg_sepal_len = df["sepal_length"].mean()
print("\nOverall average sepal length:", avg_sepal_len)
```

Use the DataFrame grouping function to determine the count by species.

Use the DataFrame grouping function to determine the average length and width of sepals and pedals by species.

```python
group_counts = df.groupby("species")["sepal_length"].count()
print("\nCount by species:")
print(group_counts)
```

Add the sum of the sepal width and length as a new column to your DataFrame.

```python
df["sepal_sum"] = df["sepal_length"] + df["sepal_width"]
print("\nAdded column 'sepal_sum':")
print(df[["sepal_length", "sepal_width", "sepal_sum"]].head())
```

Visualize petal length with a histogram

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 4))
df["petal_length"].plot(kind="hist", bins=20, edgecolor="black")
plt.title("Petal Length Histogram")
plt.xlabel("Petal length (cm)")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

Describe the distribution

```python
The histogram is clearly bimodal, meaning it has two separate peaks:
	1.	First cluster (Setosa):
	•	Very short petals
	•	Between 1.0 and 2.0 cm
	•	A tight, narrow peak → very low variation
	2.	Second cluster (Versicolor + Virginica):
	•	Much longer petals
	•	Spread from 3.0 to 7.0 cm
	•	Wider and more dispersed
	•	Contains two overlapping sub-peaks (Versicolor around ~4–5 cm, Virginica around ~5–6+ cm)

Key takeaways:
	•	The distribution is not normal; it is strongly multimodal because each iris species has different petal lengths.
	•	There is a clear separation between Setosa and the other two species.
	•	Versicolor and Virginica overlap, but Virginica tends to have the longest petals.
```

Visualize petal length and width with a "scatter_matrix"

```python
from pandas.plotting import scatter_matrix
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 6))
scatter_matrix(df[["petal_length", "petal_width"]], diagonal="hist")
plt.suptitle("Scatter Matrix: Petal Length vs Petal Width")
plt.tight_layout()
plt.show()
```

Create a new column with a rough estimate of petal area by multiplying petal length and width together.

```python
# 14) New column: rough petal area (length * width)
df["petal_area"] = df["petal_length"] * df["petal_width"]
print("\nAdded column 'petal_area':")
print(df[["petal_length", "petal_width", "petal_area"]].head())
```

Create a new DataFrame with petal areas greater than $1cm^2$.

```python
# 15) New DataFrame: petal area > 1 cm^2
df_big_petal = df[df["petal_area"] > 1.0]
print("\nRows with petal_area > 1 cm^2:")
print(df_big_petal.head())
print("Count with petal_area > 1 cm^2:", len(df_big_petal))
```
