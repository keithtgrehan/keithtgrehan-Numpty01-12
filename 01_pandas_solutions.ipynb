Load external data (abalone example from notebook)

```python
abalone_df = pd.read_csv('data/abalone.csv', header=None)
abalone_df.columns = ['sex', 'length', 'diameter', 'height', 'weight_whole', 'weight_shucked', 'weight_viscera', 'weight_shell', 'n_rings']
abalone_df.head()
```

Other tasks like filtering, grouping, etc., are demonstrated in code.

```python
# Example filter
abalone_df[abalone_df['length'] <= 0.3]
```

```python
# Example groupby
abalone_df.groupby('sex')['length'].mean()
```

```python
# Rename column
abalone_df.rename(columns={'n_rings': 'nr_rings'}, inplace=True)
```

```python
# Eval for new columns
abalone_df.eval('age = nr_rings + 1.5', inplace=True)
abalone_df.eval('weight_per_height = weight_whole / diameter', inplace=True)
```

```python
# Drop column
abalone_df.drop('nr_rings', axis=1, inplace=True)
```

```python
# Fill/drop NaNs (no NaNs in this data, but example)
abalone_df.fillna(-1, inplace=True)
abalone_df.dropna(inplace=True)
```
