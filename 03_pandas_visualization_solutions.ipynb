# Solutions: Visualization in pandas

```python
import pandas as pd
df = pd.read_csv('data/winequality-red.csv', delimiter=';')
```

**1. Plot the average amount of `chlorides` for each `quality` value (1 from Part 3).**

```python
# Group by quality and get mean chlorides, then plot as bar chart
df.groupby('quality')['chlorides'].mean().plot(kind='bar')
```

**2. Plot the `alcohol` values against `pH` values. Does there appear to be any relationship between the two?**

```python
# Scatter plot of alcohol vs pH
df.plot(kind='scatter', x='alcohol', y='pH')
# There appears to be a slight negative correlation (higher alcohol, slightly lower pH), but it's weak.
```

**3. Plot `total_acidity` values against `pH` values. Does there appear to be any relationship between the two?**

```python
# First, create total_acidity if not present (assuming it's fixed acidity + volatile acidity)
df['total_acidity'] = df['fixed acidity'] + df['volatile acidity']
df.plot(kind='scatter', x='total_acidity', y='pH')
# There is a negative relationship: higher total acidity correlates with lower pH.
```

**4. Plot a histogram of the `quality` values. Are they evenly distributed within the data set?**

```python
df['quality'].plot(kind='hist')
# No, they are not evenly distributed; most wines are rated 5 or 6.
```

**5. Plot a box plot to look at the distribution of `citric acid`.**

```python
df[['citric acid']].plot(kind='box')
```
